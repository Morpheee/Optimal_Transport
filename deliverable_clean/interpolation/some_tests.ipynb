```python
import numpy as np
import main
import matplotlib.pyplot as plt
```

```python
s = 5
img1 = np.zeros((s,s))
img1[0,0] = 1
img1[[1,0],[0,1]] = 0.5
img1[[0,1,2],[2,1,0]] = .25

img2 = np.zeros((s,s))
img2[s-1,s-1] = 1
img2[[s-2,s-1],[s-1,s-2]] = 0.5
img2[[s-1,s-2,s-3],[s-3,s-2,s-1]] = 0.25
```

```python
plt.figure(1, figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.title("image1")
plt.subplot(1,2,2)
plt.imshow(img2)
plt.title("image2")
plt.show()
```

```python
main.main(img1, img2, show=True)
```
